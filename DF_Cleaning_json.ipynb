## Format data with images

```python
import requests





def getImage(url):
    filename = url.split('/')[-1]
    r = requests.get(url, allow_redirects=True)
    open(filename, 'wb').write(r.content)
    return filename

def getName(url):
    filename = url.split('/')[-1]
    return filename

for index, row in df1.iterrows():
    filename = getName(row['image-src'])
    row['image-src'] = 'images/' + filename
    
    

```

```python
import json

rating, playingTime, playerMin, playerMax, agemin, lang,  names, image, category, description, prix   = [], [], [], [], [], [], [], [], [], [], []
for index, row in df1.iterrows():
    
    rating.append(row['note_moyenne'])
    playingTime.append(row['time'])
    
    result_player = re.findall("\d+", row['players'])
    if (len(result_player) == 2):
        playerMin.append(result_player[0])
        playerMax.append(result_player[1])
    else:
        playerMin.append(row['players'])
    
    agemin.append(row['age'])
    prix.append(row['prix'])
    lang.append(row['langue'])
    names.append(row['name'])
    image.append(row['image-src'])
   
    category.append(" ")
    description.append(row['description'])
   

games = [{"rating": r, "playingTime": s, "playerMin": pmin, "playerMax": pmax, "ageMin": agemin, "language": lang, "name": name, "image": img, "category": cat, "description": des, "price": prix} for r, s, pmin, pmax, agemin, lang, name, img, cat, des, prix in zip(rating, playingTime, playerMin, playerMax, agemin, lang, names, image, category, description, prix)]
# Printing in JSON format
print (json.dumps(games[0]))
```

```python
with open('data.txt', 'w') as outfile:  
    json.dump(games, outfile)
```
